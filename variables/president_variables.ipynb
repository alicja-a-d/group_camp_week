```python
import pandas as pd
```

```python
df = pd.read_csv('../Camp_Election.Oracle/Historical Election Results/MIT_Lab_Historical_Election_Results/1976-2020-president.csv')
df
```

```python
df = df[df['party_simplified'].str.contains('DEMOCRAT') | df['party_simplified'].str.contains('REPUBLICAN')]
df
```

```python
#result = df.groupby(['state', 'year'])[['candidatevotes', 'party_simplified']].max()
result = df.groupby(['state', 'year']).apply(lambda x: x.loc[x['candidatevotes'].idxmax()])[['party_simplified', 'candidatevotes']]
result = result.reset_index()
result
```
