```python
import pandas as pd
```

```python
result = pd.read_csv('../Camp_Election.Oracle/Historical Election Results/MIT_Lab_Historical_Election_Results/1976-2020-president.csv')
result
```

```python
result = result[result['party_simplified'].str.contains('DEMOCRAT') | result['party_simplified'].str.contains('REPUBLICAN')]
result
```

```python
#result = result.groupby(['state', 'year'])[['candidatevotes', 'party_simplified']].max()
result = result.groupby(['state', 'year']).apply(lambda x: x.loc[x['candidatevotes'].idxmax()])[['party_simplified', 'candidatevotes']]
result = result.reset_index()
result
```

```python
import pandas as pd

# Assuming 'result' is your DataFrame with columns 'state', 'year', and 'party_simplified'

# Sort the DataFrame by 'state' and 'year' to ensure proper ordering
result = result.sort_values(by=['state', 'year'])

# Create a new column to result the cumulative count of party changes
result['previous_party_changes'] = 0

# Create variables to result the last observed state and party
last_state = None
last_party = None

# Iterate over each row in the DataFrame
for idx, row in result.iterrows():
    state = row['state']
    year = row['year']
    party = row['party_simplified']
    
    # Check if we have encountered a new state
    if last_state != state:
        # Reset the cumulative count of party changes
        result.at[idx, 'previous_party_changes'] = 0
    
    else:
        # Check if the current party is different from the last observed party
        if last_party is not None and last_party != party:
            # Increment the cumulative count of party changes
            result.at[idx, 'previous_party_changes'] = result.at[idx - 1, 'previous_party_changes'] + 1
        else:
            # Copy the cumulative count from the previous row
            result.at[idx, 'previous_party_changes'] = result.at[idx - 1, 'previous_party_changes']
    
    # Update the last observed state and party
    last_state = state
    last_party = party

# Display the updated DataFrame
print(result)
```
