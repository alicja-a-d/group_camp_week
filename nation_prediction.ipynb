```python
import pandas as pd
```

```python
df = pd.read_csv('clean_TFC_data.csv')
results = pd.read_csv('Camp_Election.Oracle/Historical Election Results/MIT_Lab_Historical_Election_Results/1976-2020-president.csv')
results
```

```python
# Focus on Democrats and Republicans only 
results = results[results['party_simplified'].str.contains('DEMOCRAT') | results['party_simplified'].str.contains('REPUBLICAN')]

# Calculate the two-party total votes for each state and year
results['two_party_total'] = results.groupby(['year', 'state'])['candidatevotes'].transform('sum')


results
```

```python
results = results.drop_duplicates(subset=['state', 'year'])
df = pd.merge(df, results[['state', 'year', 'two_party_total']], on=['state', 'year'], how='left')
df
```

```python
weighted_avg = (df['pct_2p_vote'] * df['two_party_total']).groupby(df['year']).sum() / df.groupby('year')['two_party_total'].sum()
weighted_avg
```

```python
df = df[['year', 'term2', 'real_gdp_pct_growth', 'net_approval']].drop_duplicates().reset_index(drop=True)
df['pct_vote'] = weighted_avg.tolist()
df
```

```python
import statsmodels.api as sm

# Define your design matrix and your target variable 


# Specify the columns to be included in X. Ensure all are numeric
X = df[['term2', 'real_gdp_pct_growth', 'net_approval']]
y = df['pct_vote']

# Add a constant to the features to account for the intercept in the model
X_with_const = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X_with_const).fit()
print(model.summary())
```

```python
Xtest = pd.DataFrame(columns = X.columns)
Xtest = X.iloc[:1].copy()
Xtest['term2'] = 1.0
Xtest['real_gdp_pct_growth'] = 2.5
Xtest['net_approval'] = -13
Xtest
test_prediction = model.predict(Xtest)
test_prediction
```
