```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
```

```python
approval = pd.read_csv('Camp_Election.Oracle/Historical_President.Approval.Polls_Gallup.csv')
approval['net_approval'] = approval['Approving'] - approval['Disapproving']
approval
```

```python
approval['Start Date'] = pd.to_datetime(approval['Start Date'])
approval['End Date'] = pd.to_datetime(approval['End Date'])
#approval = approval[approval['Start Date'] > pd.Timestamp('2021-01-20')]
approval = approval[['Start Date', 'net_approval']]
approval.set_index('Start Date', inplace=True)
approval.sort_index(inplace=True)

approval
```

```python
model = ARIMA(approval['net_approval'], order=(2,2,1))
model_fit = model.fit()
forecast = model_fit.forecast(steps=11)
forecast
```

```python
best_aic = float('inf')
best_model = None

for p in range(3):  # Try different values of p
    for d in range(2):  # Try different values of d
        for q in range(3):  # Try different values of q
            try:
                model = ARIMA(approval['net_approval'], order=(p, d, q))
                model_fit = model.fit()
                aic = model_fit.bic
                
                if aic < best_aic:
                    best_aic = aic
                    best_model = model_fit
            except:
                continue

# Display the best-performing model and its AIC value
#print("Best ARIMA Model (p={}, d={}, q={}):".format(best_model.order[0], best_model.order[1], best_model.order[2]))
print("AIC:", best_aic)
print(best_model.summary())
```

```python
forecast = best_model.forecast(steps=11)
forecast
```

```python
aux_gdp = pd.read_csv('Camp_Election.Oracle/Historical_National.US.GDP.csv')
aux_gdp = aux_gdp[aux_gdp['Items']=='Gross domestic product']

# melt 
aux_gdp = pd.melt(aux_gdp, id_vars='Items',var_name='year',value_name='real_gdp_pct_growth')
aux_gdp = aux_gdp[['year','real_gdp_pct_growth']]
aux_gdp['real_gdp_pct_growth'] = pd.to_numeric(aux_gdp['real_gdp_pct_growth'])

# merge with df 
#df['year'] = df['year'].astype(int)  # Convert to int
aux_gdp['year'] = aux_gdp['year'].astype(int)  # Ensure this matches df
aux_gdp
```

```python
best_aic = float('inf')
best_model = None

for p in range(3):  # Try different values of p
    for d in range(2):  # Try different values of d
        for q in range(3):  # Try different values of q
            try:
                model = ARIMA(aux_gdp['real_gdp_pct_growth'], order=(p, d, q))
                model_fit = model.fit()
                aic = model_fit.bic
                
                if aic < best_aic:
                    best_aic = aic
                    best_model = model_fit
            except:
                continue

# Display the best-performing model and its AIC value
#print("Best ARIMA Model (p={}, d={}, q={}):".format(best_model.order[0], best_model.order[1], best_model.order[2]))
print("AIC:", best_aic)
print(best_model.summary())
```

```python
forecast = best_model.forecast(steps=1)
forecast
```
